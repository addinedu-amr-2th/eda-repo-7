### maker_no값 수집

```python
# 국산차 제조사명, 넘버 수집
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('headless')

url = 'https://www.bobaedream.co.kr/cyber/CyberCar.php?gubun=K'
driver = webdriver.Chrome(service=Service('../data/driver/chromedriver'), options=chrome_options)
driver.get(url)

page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')

maker_list = soup.select('div > .finder-option-list > li > button')

maker_no_list_k = []
maker_name_list_k = []
for each in maker_list:
    maker_no_list_k.append(re.search('\d+',each['onclick']).group())
    maker_name_list_k.append(each.text)

del maker_no_list_k[0]
del maker_name_list_k[0]

print(maker_no_list_k)
print(maker_name_list_k)
```

```python
# 수입차 제조사명, 넘버 수집
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('headless')

url = 'https://www.bobaedream.co.kr/cyber/CyberCar.php?gubun=I'
driver = webdriver.Chrome(service=Service('../data/driver/chromedriver'), options=chrome_options)
driver.get(url)

page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')

maker_list = soup.select('div > .finder-option-list > li > button')

maker_no_list_i = []
maker_name_list_i = []
for each in maker_list:
    maker_no_list_i.append(re.search('\d+',each['onclick']).group())
    maker_name_list_i.append(each.text)

del maker_no_list_i[0]
del maker_name_list_i[0]

print(maker_no_list_i)
print(maker_name_list_i)
```

```python
# 제조사명, 넘버 리스트 정의
maker_no_list = [maker_no_list_k,maker_no_list_i]
maker_name_list = [maker_name_list_k,maker_name_list_i]
print(maker_no_list)
print(maker_name_list)
```

### 상세페이지 url 수집

```python

href_list = []

for i in range(2): # i = 0: 국산차, 1: 수입차
    href_kor_for = []

    for j in range(len(maker_no_list[i])): # j: brand
        href_each_brand = []

        for k in range(1, 501): # k: page
            chrome_options = webdriver.ChromeOptions()
            chrome_options.add_argument('headless')

            url = 'https://www.bobaedream.co.kr/cyber/CyberCar.php?sel_m_gubun=ALL&maker_no={}&page={}&price_sel1=P0&order=S11&view_size=70'.format(maker_no_list[i][j], k)
            driver = webdriver.Chrome(service=Service('C:\Users\zxcvb\OneDrive\문서\ws\source_code\chromedriver.exe'), options=chrome_options)
            driver.get(url)

            page = driver.page_source
            soup = BeautifulSoup(page, 'html.parser')
            

            if bool(soup.find_all(class_='product-item')) == True:
                wait = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'tit.ellipsis')))
                
                tmp = soup.select('div#listCont > div.wrap-thumb-list > ul.clearfix > li.product-item > div.list-inner > div.mode-cell.title > p.tit.ellipsis > a')
                tmp = [option['href'] for option in tmp]

                href_each_brand.append(tmp)
                print('국산/수입:' + str(i) + ', 제조사:' + str(j) + ', 페이지:' + str(k))

            else:
                break

            
        
        href_each_brand = list(itertools.chain(*href_each_brand))
        href_kor_for.append(href_each_brand)

        wait = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'listCont')))
    
    href_list.append(href_kor_for)
```

### 상세페이지 접속, 크롤링

```python
car_list = []
```

```python
for i in range(2): # i = 0: 국산차, 1: 수입차
    for j in range(len(href_list[i])): # j: brand

        if href_list[i][j] == [0]: # 리스트가 비어있으면 (0이면) 건너뛰기
            continue

        for k in range(len(href_list[i][j])): # k: page
            
            chrome_options = webdriver.ChromeOptions()
            #chrome_options.add_argument('headless')
            # chrome_options.add_argument("--window-size=1920,1080")
            # chrome_options.add_argument("--start-maximized")
            # chrome_options.add_argument("--headless")
    
            url_detail = 'https://www.bobaedream.co.kr' + href_list[i][j][k]
            driver = webdriver.Chrome(service=Service('source_code\chromedriver.exe'), options=chrome_options)
            driver.get(url_detail)

            page = driver.page_source
            soup = BeautifulSoup(page, 'html.parser')

            ## 차량 정보 리스트 초기화
            car=[]
            car_brand = []
            car_name = []
            car_distance = []
            car_oil = []
            car_price = []
            car_avg_price = []
            car_color = []
            car_date = []
            car_displacement = []
            car_insurance=[]
            car_option = []
            
            
            # 차명,브랜드 수집 
            car_name_info=driver.find_element(By.CSS_SELECTOR, '#bobaeConent > div.component.page-detail > div.container-detail > div.wrap-detail-spot.js-object-fit > div.group-info > div.info-price.box.mode-basic > div.title-area > h3').text
            car_name_info=car_name_info.replace('\n','')
            car_name_info = re.split(' ',car_name_info,maxsplit=1)
            car_brand.append(car_name_info[0])
            car_name.append(car_name_info[1])

            car.append(car_brand)
            car.append(car_name)
            
            # 연식, 주행거리, 유종
            car_used_info = driver.find_element(By.CSS_SELECTOR,"#bobaeConent > div.component.page-detail > div.container-detail > div.wrap-detail-spot.js-object-fit > div.group-info > div.info-price.box.mode-basic > div.title-area > p").text
            car_used_info=re.split('\n',car_used_info)
            car_date.append(car_used_info[0])
            car_distance.append(car_used_info[1])
            car_oil.append(car_used_info[2])

            car.append(car_date)
            car.append(car_distance)
            car.append(car_oil)
                        
                        
            # 판매가, 평균시세 수집 (예외처리)
            car_price_flag=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(2) > div:nth-child(2) > div.info-price > div')
            if car_price_flag.text=='해당모델은 시세가 준비되지 않았습니다.\n동급매물 팔린매물 중고시세':
                car_price_info=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div.wrap-detail-spot.js-object-fit > div.group-info > div.info-price.box.mode-basic > div.price-area > span > b').text
                car_price.append(car_price_info)
                car_avg_price.append([np.nan,np.nan])
                car.append(car_price)
                car.append(car_avg_price)
            else: 
                car_price_info=driver.find_element(By.CSS_SELECTOR,'#meter').get_attribute('data-price')
                car_price_info=car_price_info.replace('[','').replace(']','').replace('"','')
                car_price_info=re.split(',',car_price_info)
                car_price.append(car_price_info[4])
                car_avg_price.append(car_price_info[0])
                car_avg_price.append(car_price_info[1])

                car.append(car_price)
                car.append(car_avg_price)

            # 배기량 , 색상
            car_displacement.append(driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div.detail-section.mode-half > div:nth-child(1) > div.info-basic > div > table > tbody > tr:nth-child(1) > td:nth-child(4)').text)
            car_color.append(driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div.detail-section.mode-half > div:nth-child(1) > div.info-basic > div > table > tbody > tr:nth-child(2) > td:nth-child(4)').text)
            

            car.append(car_displacement)
            car.append(car_color)
            
            
            
            ## 옵션 수집(예외처리 필요)
            car_option_flag=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(4) > div > div.detail-option-container > div > table > tbody')

            if (car_option_flag.text==''):
                car_option=[np.nan]

            else :
                # car_option1 외관
                car_option_info=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(4) > div > div.detail-option-container > div > table > tbody > tr:nth-child(2) > td > div')
                car_option.append([option.get_attribute('checked') for option in car_option_info.find_elements(By.TAG_NAME,'input')])

                # car_option2 내장
                car_option_info=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(4) > div > div.detail-option-container > div > table > tbody > tr:nth-child(3) > td > div')
                car_option.append([option.get_attribute('checked') for option in car_option_info.find_elements(By.TAG_NAME,'input')])
                
                # car_option3 안전
                car_option_info=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(4) > div > div.detail-option-container > div > table > tbody > tr:nth-child(4) > td > div')
                car_option.append([option.get_attribute('checked') for option in car_option_info.find_elements(By.TAG_NAME,'input')])
                    
                # car_option4 편의
                car_option_info=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(4) > div > div.detail-option-container > div > table > tbody > tr:nth-child(5) > td > div')
                car_option.append([option.get_attribute('checked') for option in car_option_info.find_elements(By.TAG_NAME,'input')])
                    
                # car_option5 멀티미디어
                car_option_info=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(4) > div > div.detail-option-container > div > table > tbody > tr:nth-child(6) > td > div')
                car_option.append([option.get_attribute('checked') for option in car_option_info.find_elements(By.TAG_NAME,'input')])
            
            car.append(car_option)
            
            
            ## 보험처리이력 (예외처리 필요)
            car_insurance_flag=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div:nth-child(1) > div.d-head > h4')
            car_insurance_flag.text

            if car_insurance_flag.text=='보험처리이력':
                
                car_insurance_flag2=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div:nth-child(1) > div.info-insurance > div > dl > dd')
                
                if car_insurance_flag2.text=="사고이력정보가 등록되지 않았습니다.":
                    car_insurance=[np.nan]
                else : 
                    
                    car_insurance_count=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div > div.info-insurance.mode-personal > div.top-state > dl > dd > b')
                    car_insurance_change=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div > div.info-insurance.mode-personal > div.insurance-cont > dl.flt > dd:nth-child(2)')
                    car_insurance_sago=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div > div.info-insurance.mode-personal > div.insurance-cont > dl.flt > dd:nth-child(4)')
                    car_insurance_mycarbroken=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div > div.info-insurance.mode-personal > div.insurance-cont > dl.frt > dd:nth-child(2)')
                    car_insurance_youcarbroken=driver.find_element(By.CSS_SELECTOR,'#bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div > div.info-insurance.mode-personal > div.insurance-cont > dl.frt > dd:nth-child(4)')
                    
                    
                    #bobaeConent > div.component.page-detail > div.container-detail > div:nth-child(5) > div:nth-child(1) > div.info-insurance > div > dl > dd
                    ## 텍스트 분리
                    # 차량번호 / 소유자변경
                    car_insurance_change_=re.split(' ',car_insurance_change.text)
                    car_insurance_change_car=car_insurance_change_[0][0]
                    car_insurance_change_owner=car_insurance_change_[2][0]
                    
                    # 전손 / 침수전손 / 침수분손 / 도난
                    car_insurance_sago_=re.split(' ',car_insurance_sago.text)
                    car_insurance_sago_total_loss=car_insurance_sago_[1]
                    car_insurance_sago_Flooding_total=car_insurance_sago_[5]
                    car_insurance_sago_Flooding_fraction=car_insurance_sago_[9]
                    car_insurance_sago_theft=car_insurance_sago_[-1]
                    
                    # 내차피해 (횟수/금액) / 상대차피해 (횟수/금액)
                    car_insurance_mycarbroken_=re.split(' ',car_insurance_mycarbroken.text)
                    car_insurance_mycarbroken_count=car_insurance_mycarbroken_[0][0]
                    car_insurance_mycarbroken_money=car_insurance_mycarbroken_[1][1]
                    car_insurance_youcarbroken_=re.split(' ',car_insurance_youcarbroken.text)
                    car_insurance_youcarbroken_count=car_insurance_youcarbroken_[0][0]
                    car_insurance_youcarbroken_money=car_insurance_youcarbroken_[1][1]
                    
                    car_insurance.append([
                        car_insurance_count.text,
                        car_insurance_change_car,
                        car_insurance_change_owner,
                        car_insurance_sago_total_loss,
                        car_insurance_sago_Flooding_total,
                        car_insurance_sago_Flooding_fraction,
                        car_insurance_sago_theft,
                        car_insurance_mycarbroken_count, 
                        car_insurance_mycarbroken_money,
                        car_insurance_youcarbroken_count,
                        car_insurance_youcarbroken_money])
                    
                    
            else:
                print('보험처리이력 없음')
                car_insurance.append([np.nan])
                
                
                
            car.append(car_insurance)
            
            print(car)   
            car_list.append(car)        
            print('국산/수입:' + str(i) + ', 제조사:' + str(j) + ', 페이지:' + str(k) + ', 차량:' + str(car_name_info[1])+ ', 브랜드:' + str(car_name_info[0]))
```
